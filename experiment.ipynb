```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
df_mk1 = pd.read_csv('mk1.csv', sep=',')
print(df_mk1)
```

```python
linear = LinearRegression()
x1 = df_mk1[["Leader's Price"]]
y1 = df_mk1[["Follower's Price"]]
linear.fit(x1, y1)
score1 = linear.score(x1, y1)
print(score1)

plt.plot(df_mk1[["Leader's Price"]], linear.predict(df_mk1[["Leader's Price"]]), color = 'r',label = 'predicition')
plt.scatter(df_mk1[["Leader's Price"]], df_mk1[["Follower's Price"]], color = 'b',label = 'original_data')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
model = LinearRegression()
x1 = df_mk1[["Leader's Price"]]
y1 = df_mk1[["Follower's Price"]]

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1)

model.fit(x1_train, y1_train)
train_score = model.score(x1_train, y1_train)
test_score = model.score(x1_test, y1_test)
print("train score: " + str(train_score))
print("test score: " + str(test_score))
```

```python
from sklearn.model_selection import train_test_split
model = DecisionTreeRegressor()
x1 = df_mk1[["Leader's Price"]]
y1 = df_mk1[["Follower's Price"]]

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1)

model.fit(x1_train, y1_train)
train_score = model.score(x1_train, y1_train)
test_score = model.score(x1_test, y1_test)
print("train score: " + str(train_score))
print("test score: " + str(test_score))

plt.plot(df_mk1[["Leader's Price"]], model.predict(df_mk1[["Leader's Price"]]), color = 'r',label = 'predicition')
plt.scatter(df_mk1[["Leader's Price"]], df_mk1[["Follower's Price"]], color = 'b',label = 'original_data')
plt.show()
```
